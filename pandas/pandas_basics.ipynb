## **Pandas Series** --> can hold different datatypes like lists (Unlike ndarrays). And also can name the indices of Pandas series elements.[Same as Dictionary]

```python
import pandas as pd

x = pd.Series(data=[5,4.6,"Tasty!"],index=["milk","chips","chocolate"])          #  Syntax: pd.Series(data,index)

print(x)                           # gives in a single colomn of elements
print(x[0])
print(x["milk"])
print(x[-3])
```

#### **Ways to Access:**
####**1) Index**
####**2) List of Index**
####**3) Slicing**

```python
import pandas as pd

x = pd.Series(data=[5,4.6,"Tasty!"],index=["milk","chips","chocolate"])

print(x["milk"])
print(x[["chips","milk"]])             # can use user-defined list of indices
print()
print(x[[0,1]])                        # can use predefined list of indices
print()
print(x[1:2])                          # can slice the colomn --> endpoint=inclusive (like linspace)
print()
print(x["chips":"chocolate"])          # can slice the colomn --> endpoint=inclusive only when using user-defined indices (like linspace)
```

```python
import pandas as pd

x = pd.Series(data=[5,4.6,"Tasty!"],index=["milk","chips","chocolate"])

print(x.loc[["chocolate","milk"]])
print()
print(x.iloc[[2,1]])
```

#### **Pandas Series are mutable like ndarrays and lists i.e, gives view instead of copy**

```python
import pandas as pd

y = pd.Series(data=[5,4.6,"Tasty!","joy",88],index=["milk","chips","chocolate","stick","dawg"])

print(y.drop("chocolate"))            # does a temporary deletion like sorted() --> temporary sort
print()
print(y)

print()
y = y.drop("chocolate")              # to delete permantly, we can override it
print(y)
print()
#            [or]
y.drop("stick",inplace=True)          # [or] we can use "inplace" --> **saying make this change inplace, which is permanent change**
print(y)
```

### **Deleting the Pandas DataFrame**

```python
import pandas as pd

x = [{"apple":40,"banana":70,"grapes":89},{"apple":74,"grapes":56,"tomato":78}]

j = pd.DataFrame(x,index=["person1","Person2"])

j.pop("grapes")                                                 # x.pop() --> deletes columns  => permanent change
j.drop(["apple","banana"],axis=1,inplace=True)                  # x.drop([column_name/row_name],axis,inplace) => deletes both rows and columns =>
j.drop(["person1"],axis=0,inplace=True)                         # inplace=True -->Permanent change, default --> Temporary change
print(j)
```

#### **SUMMARY: x.pop(column_name) --> deletes only columns**

### **Dealing with NaN values**

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":23,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":35,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["store 1","store 2","store 3"])
print(d)

print()
print(d.isnull())                                   # x.isnull() --> checks if "NaN" is present and returns Bool

print()
print(d.isnull().sum())                             # sum() --> sums the total of "NaN" in each column's values   --> True = 1 , False = 0
print(d.isnull().sum().sum())                       # As sum() adds values in columns, it adds total no. of "NaN"  --> gives no. of "NaN" in the entire Data
```

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":23,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":35,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["store 1","store 2","store 3"])

print(d["shirts"].isnull())                       # applys .isnull() for "shirts" column and gives bool of columns
print()
print(d[["shirts","suits"]].isnull())               # can do above process for multiple columns

print()
print(d["shirts"].isnull().sum())                   # gives sum of values (True=1,False=0) of column

print()
print(d[["shirts","suits"]].isnull().sum())         # gives sum of "NaN" in each columns and represent it in 1 column
print(d[["shirts","suits"]].isnull().sum().sum())   # gives sum of column of above derived column

print()
print(d[d["shirts"].isnull()])                      # shows the rows where "shirts" has NaN value --> As d[True] => executes
```

### **Deleting and Replace the NaN values**

```python
# Deleting: used when 90%..95% of rows/columns are NaN value to avoid Miscalculation
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":23,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":35,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["case 1","case 2","case 3"])
print(d)

print()
print(d.dropna(axis=0))                     # d.dropna() --> used to drop rows/columns which has NaN values

print()
print(d.dropna(axis=0,how="all"))           # how= --> deletes only if "all" are NaN values
```

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":23,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":35,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["case 1","case 2","case 3"])
print(d)

print()
print(d.dropna(axis=0,how="all",subset=["shirts","glasses","shoes"]))   # deletes a row where column "shirts","glasses","shoes" are all NaN value -->result: as "glasses" != NaN
print()
print(d.dropna(axis=0,how="all",subset=["shirts","suits"]))
print()
print(d.dropna(axis=1,how="all",subset=["case 1"]))                     # likewise can do for columns deletion
print()
print(d.dropna(axis=1,how="all",subset=["case 3"]))
```

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":23,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":35,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["case 1","case 2","case 3"])
print(d)

print()
print(d.dropna(axis=0,how="any"))
print()
print(d.dropna(axis=0,how="any",subset=["glasses"]))
print()
print(d.dropna(axis=1,how="any",subset=["case 1","case 3"]))
```

#### **SUMMARY: d.dropna(axis,how,subset) --> Temporary actually (out of place)**
#### **d.dropna(axis,how,subset,inplace=True) [or] d = d.dropna(axis,how,subset) --> Permanent**
********************************************************************************
####**how= :**
#### **"all" => every value should should be NaN**
####         **"any" => anyone 1 value is NaN**
********************************************************************************
#### **subset= : Every subset mentioned should have NaN value otherwise it does not perform well. --> only for "all"**
#### **Even 1 has NaN value in row/column it gets executed and deleted**

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":35,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":45,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["case 1","case 2","case 3"])
print(d)

print()
print(d.fillna(0))                                  # d.fillna() -->  fills the NaN values with other values

print()
print(d["shirts"].median())                        # In place of NaN mostly we fill it with mode or median. But for mean, we need to learn about outliers.
print()
print(d["shirts"].mode())
```

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":35,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":45,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["case 1","case 2","case 3"])
print(d)

print()
print(d.fillna(method="ffill",axis=0))                       # "ffill" --> fills the NaN value with previous row(axis=0)/column(axis=1) and if there is Nothing, then no change
print()
print(d.fillna(method="ffill",axis=1))
```

```python
import pandas as pd

i = [{"bikes":94,"pants":48,"watches":76,"shirts":35,"shoes":45,"suits":67},
     {"watches":69,"glasses":84,"bikes":47,"pants":63,"watches":90,"shirts":45,"shoes":85,"suits":74},
     {"bikes":70,"pants":30,"watches":49,"glasses":67,"shoes":56}]

d = pd.DataFrame(i,index=["case 1","case 2","case 3"])
print(d)

print()
print(d.fillna(method="backfill",axis=0))                  # "backfill" --> fills the NaN with successive row(axis=0)/column(axis=1) and if there is Nothing, then no change
print()
print(d.fillna(method="backfill",axis=1))
```

#### **SUMMARY: d.fillna(method,axis) --> Temporary actually (out of place)**
#### **d.fillna(method,axis,inplace=True) [or] d = d.fillna(method,axis) --> Permanent**
********************************************************************************
#### **method= :**    
#### **"ffill" => fills the NaN value with previous row(axis=0)/column(axis=1) and if there is Nothing, then no change.**
#### **"backfill" => fills the NaN with successive row(axis=0)/column(axis=1) and if there is Nothing, then no change.**

###**Outliers (need to be 3 standard deviations away from mean, then it is called outliers in Normal Distribution): There are those which incorrectly impact the data while doing mean of data, which cause wrong result. They should not be included while doing mean.**
####**Eg: we see dress malls make a threshold of price to give free gifts. so, to calculate this threshold, we need to calculate mean of expenditure of each person. like say, let most people are at 4k to 6k. there are some people who are 1Lkh as due to some function in their homes. so, while calculating mean these 1Lkh people which are called Outlies may disrupt the result. so, we exclude them.**
