```python
# PROBLEM
# immutable: cannot modify the value.(but can be modified when its copy is created) eg: strings (when "=" it makes the copy of the original value
# i.e if changed in duplicate, original remains same [different locations])
# mutable: can modify the value. eg: lists, dic.. (when "=" it make the view of the original value i.e if changed in duplicate, original changes [same location])

a = "god"
b = a
print(id(b))    # location of b
print(b)
b = b + "less"
print(b)
print(id(b))
print(id(a))      # original didn't change for immutable data, when its copy is modified

print()

k = ["hello", "zero", "kaifs", "gsjjs", "fjjsnk"]
d = k
d[1] = "jimmy"
print(k)           # original also changed, when modified its copy.
print(id(k))
print(id(d))
```

```python
# SOLUTION: making a copy for mutable datatype
# shallow copy: only for single lists

g = ["hello", "zero", "kaifs", "gsjjs", "fjjsnk"]
s = g.copy()                                         # created copy of it which is in different memory location (which can be said that it is a proper copy)
s[1] = "peter"
print(s)
print(g)
print(id(s))
print(id(g))
```

```python
# deep copy: for nested lists

import copy

g = ["hello", "zero", "kaifs", ["gsjjs", "fjjsnk"]]
s = copy.deepcopy(g)
s[3][0] = "piggy"
print(s)
print(g)
```
