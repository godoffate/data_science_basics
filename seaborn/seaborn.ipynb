```python
import seaborn as sns

x = sns.load_dataset("tips")  

print(x[["total_bill","tip","size"]].corr())  # corr() --> method is only done to datasets of no strings. If there are strings, we need to add columns of no strings manually.

sns.heatmap(x[["total_bill","tip","size"]].corr(),annot=True,cmap="crest")
```

**corr() --> method is only done to datasets of no strings. If there are strings, we need to add columns of no strings manually. [Due to python version, it will happen. But for some other python functions, it filters the strings and give us the corr()]**

heatmap() graph syntax: **heatmap(x.corr(),annot="bool",cmap="color_maps")  --> For a large dataset, if we want to see coorelation, we cannot analyse properly in the tables so, we use heatmap spcifically to identify the coorelation of the columns.**

annot= --> annotate the coorelation up to 2 decimal & show it in the heatmap.
cmap=  --> can give other pattern of color map. If not mentioned, it is taken as default.
eg: crest,coolwarm, etc..

**Distribution Plots: analysis of columns/variables which has no. values.**
1) joinplot
2) pairplot
3) distplot

```python
import seaborn as sns

r = sns.load_dataset("tips")

sns.jointplot(x="tip",y="total_bill",data=r,kind="hex")   # represents central part in hexagram design
```

**Univariate Analysis:** A plot of a representation of analysis of single feature/variable. eg: histogram --> takes only 1 argument i.e, list of elements.
**Bivariate Analysis:** A plot of represenation of analysis/relation of two features/variables.
**Multivariate Analysis:** A plot of representation of analysis/relation of multi features/variables.

jointplot() grapf syntax: **jointplot(x="x_coordinate",y="y_coordinate",data="where_x,y_are_coming_from",kind="how_plot_in_central_part_should_be_represented")**

kind= --> If no kind= (default), then they are represented as dots.

Jointplot: It is a Bivariate analysis of data which can represent the relation of 2 variables/features in the plot. 

```python
import seaborn as sns

r = sns.load_dataset("tips")

sns.jointplot(x="tip",y="total_bill",data=r,kind="reg")  # represent central part with a regression line(i.e best fit line amongst the dots)
```

**Regression line:** It gives the accurate visual line among all the scattered dots. Through it, if we know "y" we can interpret the "x" and vice versa.**It can be wrong as well but in linear regression, we use it to check the accuracy of the model of how much times it is correct.**

```python
import seaborn as sns

r = sns.load_dataset("tips")
 
sns.pairplot(r)          # gives the relations of all variables.
```

```python
import seaborn as sns

r = sns.load_dataset("tips")

sns.pairplot(r,hue="day")      # can know at saturdays we get huge tips for high total_bills.
```

pairplot() plot syntax: **pairplot(data,hue="can_divide_into_gropes")**

hue= --> can divide the data according to argument given.[need to be catogorical/string column]

Pairplot: Also called Scatter plot. It is Multivariate analysis of data which represents the relations of multiple features/variables among eachother. Here ignore the diagonal once as they are relation of eachother variable.[efficient if you wanted to see relations of variabkes in 1 page]

**IMPORTANT: Despite not using corr() and stating variables(x=,y=) , this pairplot() can identify the varibles which are no. and which are strings.**

```python
import seaborn as sns

r = sns.load_dataset("tips")

sns.distplot(r["tip"],kde=False,bins=100)
```

distplot() graph syntax: **distplot(column,kde="bool",bins="no_of_section_division")**

kde= --> shows the line in the graph.
bins= --> is in how many sections we want to see the data. eg: range=[1,10] then, if bins=100 --> 10-1/100 = 0.09 length for each bins. In above, there are 100 bins/sections.

Distplot: It tells the distribution of the columns features/variables i.e, same as Histogram.

**Catagories Plots: analysis of columns/variables which has string values.**
1) countplot
2) barplot
3) boxplot
4) violinplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset("tips")

plt.subplot(2,1,1)
sns.countplot(x="sex",data=a)           # represent "sex" in x-axis

plt.subplot(2,1,2)
sns.countplot(y="day",data=a)           # represent "sex" in y-axis
```

countplot() graph syntax: **countplot(x/y="x/y_coordinate",pallete="color_map",data="where_x/y_are_coming_from")**
Here, can mention only x or only y but not both.

Countplot: It counts the string value columns, how many times each catogory repeated. 

```python
import seaborn as sns

a = sns.load_dataset("tips")

sns.barplot(x="sex",y="total_bill",data=a,hue="smoker")   # represents smoker male&female and non-smoker male&female and their total_bill
```

barplot() graph syntax: **barplot(x="x_coordinate",y="y_coordinate",data="where_x/y_are_coming_from",hue="can_divide_into_gropes")**

hue= --> can divide the data according to argument given.[need to be catogorical/string column] **Here, 2 features/string column can be compared against total_bill.**

**Barplot:** we can take 2 variables x,y one string column and one no. column and relate the catogories in string column with the no. column.

```python
import seaborn as sns

a = sns.load_dataset("tips")

sns.boxplot(x="day",y="total_bill",data=a,hue="smoker",orient="v",palette="rainbow")
```

boxplot() graph syntax: **boxplot(x="x_coordinate",y="y_coordinate",data="where_x/y_are_coming_from",hue= "can_divide_into_groupes",orient="box_orientation",palette="color_map")**

hue= --> can divide the data according to argument given.[need to be catogorical/string column]
orient= --> same function as in vert= in heatmap() i.e, can shows boxes horizontal or vertical(v). Default: horizontal
palette= --> same function as cmap in heatmap() i.e, combination of colors.

Boxplot: same as in matplotlib.pyplot , where it can show outliers.

```python
import seaborn as sns

a = sns.load_dataset("tips")

sns.violinplot(x="total_bill",y="day",data=a,palette="rainbow")
```

violinplot() graph syntax: **violinplot(x="x_coordinate",y="y_coordinate",data="where_x/y_are_coming_from",palette="color_map")**

Violinplot: It is nothing but the combination of boxplot() and distribution of data(standard deviation)

**NOTE: In seaborn, we use a dict_of_elements/list_of_dict as the form of data. Therefore, we use pandas --> efficient for dictionaries. Unlike in matplotlib.pyplot, we use numpy i.e, list_of_elements as data.**
