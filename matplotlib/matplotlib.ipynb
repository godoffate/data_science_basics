```python
# To find the Outliers, the most familier way is to use boxplots

import matplotlib.pyplot as plt
import numpy as np

s = [np.random.normal(0,std,size=(100)) for std in range(1,4)]

plt.boxplot(s,vert=True,patch_artist=True)    # --> vert= -> vertically_represent, patchartist= -> fill color
```

boxplot graph syntax: **boxplot(list_of_elements,vert="bool",patch_artist="bool")**

Here, the box represents, from bottom is 0% --> 25% --> 50% --> 75% --> 100% [percentile]. If there are dots outside this boxes, they are identified as Outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

l = "Python","C++","Ruby","Java","React.js","Express","MongoDB"
x = [213,322,488,426,893,542,479]
c = ["gold","yellowgreen","lightcoral","lightskyblue","pink","black","purple"]
e = (0,0,0.7,0.2,1,0.85,0.4)                                  # explode --> list of distances for exploding

z = list(zip(l,x))

plt.subplot(2,1,1)
plt.pie(x,explode=e,labels=l,colors=c,autopct="%1.1f%%",shadow=False,labeldistance=1.5,pctdistance=1.1) 

plt.subplot(2,1,2)
plt.pie(x,explode=e,labels=z,colors=c,shadow=False,labeldistance=0.5)
```

pie() graph syntax: **pie(data,explode="list_of_distances_to_explode",labels="list_of_names_of_slices",colors="list_of_colors_of_slices",autopct=%(percentage_format)%,shadow="bool",labeldistance="label_distance_wrt_slices",pctdistance="percentile_distance_wrt_slices")**

autopct= --> makes the given list of elements into percentiles. 
Here, %(1.1f%)% --> number means it depicts till 1 decimal place and % means it should have at the end.

It is recommended to keep the default positions for percentile and labels i.e, don't write labeldistance,pctdistance.

**NOTE: In matplotlib.pyplot, we use a list_of_elements as the form of data. Therefore, we use numpy --> efficient for lists.**
**And for the dict_of_elements as a form of data, we can use pandas --> effecient for dictionary.**
